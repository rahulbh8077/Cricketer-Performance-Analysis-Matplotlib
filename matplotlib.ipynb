```python
# a simple matplotlib example has been added below to show how to use matplotlib to plot a simple line graph.
sales=[100, 75, 200, 121, 101]
year=[2015, 2016, 2017, 2018, 2019]
plt.plot(year, sales)
plt.title("Sales Over Years")
plt.xlabel("Year")  
plt.ylabel("Sales")
```

# plotting the and reading players Data

```python
batsmen=pd.read_csv("sharma-kohli.csv")
batsmen
```

```python
# plotting the data of batsmen using matplotlib of Virat Kohli
plt.plot(batsmen['index'], batsmen['V Kohli'])
         
```

```python
# plotting the data of batsmen using matplotlib of RG Sharma
plt.plot(batsmen['index'], batsmen['RG Sharma'])
```

# Comparision In Rohit Vs Virat Score SeasonWise.

```python
#we can differnetiate the two lines by using different colors and line widths. The following code shows how to do that.
plt.plot(batsmen['index'], batsmen['V Kohli'],color='blue',linewidth=2)
plt.plot(batsmen['index'], batsmen['RG Sharma'],color='black')
plt.title("Batsmen Performance Over Years")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

# Use of Markers

```python
# usually the D represents a diamond marker and the o represents a circle marker.
plt.plot(batsmen['index'], batsmen['V Kohli'],color='blue',linewidth=2,marker='D')
plt.plot(batsmen['index'], batsmen['RG Sharma'],color='black',marker='o')
plt.title("Batsmen Performance Over Years")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# Now we are using the Labels To understand the data of batsmen using matplotlib of Virat Kohli and RG Sharma
plt.plot(batsmen['index'], batsmen['V Kohli'],color='blue',linewidth=2,marker='D',label='V Kohli')
plt.plot(batsmen['index'], batsmen['RG Sharma'],color='black',marker='o',label='RG Sharma')
plt.title("Batsmen Performance Over Years")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
# legend function is used to show the labels of the two lines in the graph.
plt.legend()
```

# Graph enabled with Grids

```python
plt.plot(batsmen['index'], batsmen['V Kohli'],color='blue',linewidth=2,marker='D',label='V Kohli')
plt.plot(batsmen['index'], batsmen['RG Sharma'],color='black',marker='o',label='RG Sharma')
plt.title("Batsmen Performance Over Years")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
# legend function is used to show the labels of the two lines in the graph.
plt.legend()
plt.grid()
# grid function is used to show the grid lines in the graph and it is used to make the graph more readable and understandable.
```
