# Scatter Plot 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading a CSV file
df=pd.read_csv('batter.csv')
df
```

```python
df.shape
```

```python
# printing the first TOP 50 Batsmen
df.head(50)
df
```

```python
plt.scatter(df['avg'], df['strike_rate'], color='blue')
plt.title('Scatter plot of Average vs Strike Rate')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.legend(['Batsmen'])
# legend automaticcally assigned where the space is free in graph.
```

```python
# scatter plot of average > 30 with strike rate
plt.scatter(df['avg']>30,df['strike_rate'],color='red',label='batsmen')
plt.title('scatter plot of average > 30 with strike rate')
plt.xlabel('Average > 30')
plt.ylabel('Strike Rate')
plt.legend(['Batsmen'])
```

```python
# Scatter plot for runs>5000 Vs Strike Rate
plt.scatter(df['runs']>5000,df['strike_rate'],color='green',label='batsmen')
plt.title('scatter plot of runs > 5000 with strike rate')
plt.xlabel('Runs > 5000')
plt.ylabel('Strike Rate')
plt.legend(['Batsmen'])
```
