```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip')
# this is basically a axis level function, so we can use it to plot on a specific axis
```

```python
sns.relplot(data=df, x='total_bill', y='tip', kind='scatter',hue='sex',style='time')
# a figure shaped function and majourly in  sq shaped format.
# by using hue we can add a color to the scatter plot and by using style we can add a shape to the scatter plot.
# by using style we cam add a shape to the scatter plot and by using hue we can add a color to the scatter plot.
```

# line plot

```python
sns.lineplot(data=df, x='total_bill', y='tip',hue='sex')
#lineplot is basically a continous form of data to plot graphs.
```

```python
sns.relplot(data=df, x='total_bill', y='tip',hue='sex',kind='line')
# it is basically a good practise set.
```

# Histogram Plotting

```python
# histogram is basically a continous form of data to plot graphs.
sns.histplot(data=df,x='day',hue='sex')
```

```python
# infomative graph showing the majour distribution of the data.
sns.histplot(data=df,x='total_bill',hue='sex')
```

# KDE Plot

```python
# it generally shows the distribution of the data in a density format
sns.kdeplot(data=df,x='total_bill',hue='sex')
```

```python
# filling the area under the curve of the density plot.
sns.kdeplot(data=df,x='total_bill',hue='sex',fill=True)
```

# Rug Plot

```python
sns.kdeplot(data=df,x='total_bill',hue='sex')
sns.rugplot(data=df,x='total_bill',hue='sex')
```
