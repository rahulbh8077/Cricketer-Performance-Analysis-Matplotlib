```python
# creating a Distribution Plot
sns.displot(df['total_bill'], bins=30, kde=True, color='blue')
# bins is defined as the no of bars used in histogram.
# kde is majourly the kernal density distributer
# graph is plotted successfully.
```

```python
# Reminder to note that the displot is only used in one variable system.
sns.displot(df['total_bill'],df['tip'] ,bins=30, kde=True, color='blue')
```
